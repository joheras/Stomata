Once the above files are downloaded, we can load the functionality to detect stomata in images, and some additional libraries that are necessary for uploading your images.


```
from imageDetection import *
from google.colab import files
import cv2
```

## Loading the images
You can upload your own images by executing the following cell. After executing that cell a Browse button to select an image from your computer will appear.

```
uploaded = files.upload()
for fn in uploaded.keys():
       imagePath = fn
```

## Detection

In this step, we invoke the *predict* function that returns the uploaded image with the detected stomata. 


```
predict(imagePath)
```

Now, we show the result inside the notebook.

```
image = cv2.imread(imagePath.split('.')[0]+'_predict.jpg')
showImage(image)
```

It is also possible to download the image containing the detections by executing the following cell.


```
files.download(imagePath.split('.')[0]+'_predict.jpg')
```
